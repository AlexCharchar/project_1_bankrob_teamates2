```python
def haversine_vectorize(x1, y1, x2, y2):
    x1, y1, x2, y2 = map(np.radians, [x1, y1, x2, y2])
    
    new_x = x2 - x1
    new_y = y2 - y1
    
    haver_formula = np.sin(new_y/2.0)**2 + np.cos(y1) * np.cos(y2) * np.sin(new_x/2.0)**2
    dist_ = 2 * np.arcsin(np.sqrt(haver_formula))
    km =  6373 * dist_ # to get dist in km
    return km

haversine_vectorize(0.022189, -0.001545, 0.043129, -0.010346)
```
